```python
import numpy as np
# bmp 图片后缀
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.neighbors import KNeighborsClassifier
```

```python
digit = plt.imread('./1-KNN/exercise/data/3/3_1.bmp')
# 28 * 28 很小
# 首先将图片放大显示
plt.figure(figsize=(2,2))
plt.imshow(digit,cmap = 'gray')
```

```python
digit
```

```python
# 批量获取数据
data = []
target = []

for i in range(10):
    for j in range(1,501):
#         plt.imread('./data/3/3_200.bmp')
#         digit是二维的数组，黑白
        digit = plt.imread('./data/%d/%d_%d.bmp'%(i,i,j))
        data.append(digit)
        
#         目标值
        target.append(i)
```

```python
# data 和target 5000个数据
len(data)
```

```python
# 将机器学习的数据转换成ndarry，操作起来比较方便
# ndarray 提供了很多方法
data = np.array(data)
target = np.array(target)

print(data.shape,target.shape)
```

```python
index = np.random.randint(0,5000,size = 1)[0]

print('该索引所对应的目标值： ',target[index])

digit = data[index]
plt.figure(figsize=(2,2))
plt.imshow(digit,cmap = 'gray')
```

```python
seed = np.random.randint(0,5000,size = 1)[0]

# 指明随机数的种子，打乱顺序
np.random.seed(seed)
np.random.shuffle(data)

# 指明的种子和上面的一样，打乱顺序的规则就和上面一样
np.random.seed(seed)
np.random.shuffle(target)
```

```python
# 验证一下，顺序是否匹配
index = np.random.randint(0,5000,size = 1)[0]

print('该索引所对应的目标值： ',target[index])

digit = data[index]
plt.figure(figsize=(2,2))
plt.imshow(digit,cmap = 'gray')
```

```python
target[:500]
```

```python
knn = KNeighborsClassifier(n_neighbors=20)
```

```python
# 最后保留50个数据作为预测数据集

# 训练数据
X_train,y_train = data[:4950],target[:4950]


# 测试数据
X_test,y_test = data[-50:],target[-50:]
```

```python
X_train.shape
```

```python
# ValueError: Found array with dim 3. Estimator expected <= 2.
# 所给的数据三维，算法只能接受二维的数据
knn.fit(X_train,y_train)
```

```python
28*28
```

```python
X_train = X_train.reshape((4950,784))

# 正着数像素784像素点，倒着数-1
X_test = X_test.reshape((50,-1))
```

```python
# 第一步进行训练
knn.fit(X_train,y_train)
```

```python
# 第二步，使用算法，进行预测
y_ = knn.predict(X_test)

print('真实数据：',y_test)
print('预测数据： ',y_)
```

```python
# 50个数据
knn.score(X_test,y_test)
```

```python
knn.score(X_train,y_train)
```

```python
# 可视化，将50张绘制出来

plt.figure(figsize=(2*5,3*10))

for i in range(50):
    
#     10行5列
#     子视图，在每一个子视图中绘制图片
    subplot = plt.subplot(10,5,i+1)
    
#     最后50张图片在 X_test中
#     !!! 像素点需要reshape成图片形状
    subplot.imshow(X_test[i].reshape((28,28)))
    
#     添加标题，True：0
#               Predict：0
#     y_test ----真实
#     y_  ------预测
    t = y_test[i]
    p = y_[i]
    subplot.set_title('True: %d\nPredict:%d'%(t,p))
```
