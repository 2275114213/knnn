```python
import numpy as np
import matplotlib .pyplot as plt 
from sklearn.neighbors import KNeighborsClassifier
```

```python
digit = plt.imread('./data/4/4_108.bmp')
plt.figure(figsize=(2,2))
plt.imshow(digit)
```

```python
#批量获取数据
data = []
target = []
for i in range(10):
    for j in range(1,501):
#         digit = plt.imread('./data/4/4_108.bmp')
        
        digit = plt.imread('./data/%d/%d_%d.bmp'%(i,i,j))
        data.append(digit)
        target.append(i)
```

```python
#data和target都是5000个数据
# len(data)
# len(target)
```

```python
#将机器学习的数据转换为ndarray，操作起来比较方便
data = np.array(data)
target = np.array(target)
print(data.shape,target.shape)
```

```python
index = np.random.randint(0,5000,size=1)[0]
print('该索引对应的目标值',target[index])
digit = data[index]
plt.figure(figsize=(2,2))
plt.imshow(digit,cmap='gray')
```

```python
seed = np.random.randint(0,5000,size=1)[0]
np.random.seed(seed)
np.random.shuffle(data)

np.random.seed(seed)
np.random.shuffle(target)
target[:500]
```

```python
knn = KNeighborsClassifier(n_neighbors=20)
```

```python
#最后保留五十个数据作为预测数据
X_train,y_train = data[:4950],target[:4950]

X_test,y_test = data[-50:],target[-50:]
```

```python
X_train.shape
```

```python
#ValueError: Found array with dim 3. Estimator expected <= 2.
#所给数据为三维，但是算法只能接受二维
knn.fit(X_train,y_train)
```

```python
X_train = X_train.reshape(4950,784)
#像素点一共784个，倒着数为-1
X_test = X_test.reshape(50,-1)
```

```python
knn.fit(X_train,y_train)
```

```python
y_ = knn.predict(X_test)
print('真实:',y_test)
print('预测:',y_)
```

```python
#50个数u
knn.score(X_test,y_test)
```

```python
knn.score(X_train,y_train)
```

```python
#可视化，将五十张图片绘制出来
plt.figure(figsize=(2*5,2.7*10))
for i in range(50):
    #10行5列
    #子视图
    #图片需要reshape回原来的形状
    subplot = plt.subplot(10,5,i+1)
    #最后五十张图片在X_test中
    subplot.imshow(X_test[i].reshape(28,28))
    #添加标题true  predict
    #y_test     真实
    #y_预测
    t = y_test[i]
    p = y_[i]
    subplot.set_title('True:%d\nPredict%d'%(t,p))
```
